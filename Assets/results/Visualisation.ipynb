```python
import matplotlib.pyplot as plt
import pandas as pd

if __name__ == '__main__':
    path = "GenerationResults.txt"
    column_header = ["Chromosome id", "Generation Time (ms)","Score", "Generation"]
    df = pd.read_csv(path, header = None, names=column_header)
    fig1 = df.plot.scatter(x=column_header[3], y=column_header[1], title="Generation vs Generation Time").get_figure()
    fig2 = df.plot.scatter(x=column_header[1], y=column_header[2], title="Generation Time vs Score").get_figure()
#     plt.show(block=True)
    fig1.savefig('./Graphs/Generation_vs_Generation_Time_v2.png')
    fig2.savefig('./Graphs/Generation_Time_vs_Score_v2.png')
```
